```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("dataset/Covid.csv")
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['cough'])
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop('covid',axis=1),df['covid'],test_size=0.3,random_state=42)
```

```python
X_train.isnull().sum()
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
transformer=ColumnTransformer( transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['None','Mild','Moderate','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])
],
    remainder='passthrough')
```

```python
X_train_transformed=transformer.fit_transform(X_train)
X_test_transformed=transformer.transform(X_test)

xt=pd.DataFrame(X_train_transformed)
```
