# Normalization

    scale the data btw a specific range (generaly [0,1])

# Kinds of Normalization

    MinMaxScaling()
    Mean Normalization()
    Max Absolute Scaling()

## MinMaxNormalization
    formula:
                Xi'= (Xi-Xmean)/(Xmax-Xmin)

    * might change the shape a bit 

# Code Eg

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("dataset/Social_Network_Ads.csv")
df.sample(10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(
                                                df.drop("Purchased",axis=1),
                                                df["Purchased"],
                                                test_size=0.3,
                                                random_state=0
)

X_train
```

# MinMaxScaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
MinMaxScaler=MinMaxScaler()
```

```python
MinMaxScaler.fit(X_train)
```

```python
x1_train=pd.DataFrame(np.round(MinMaxScaler.transform(X_train),3),columns=X_train.columns)
x1_test=pd.DataFrame(np.round(MinMaxScaler.transform(X_test),3),columns=X_test.columns)
```

```python
x1_test
```

# Visualizing 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x=X_train.Age,y=X_train.EstimatedSalary,hue=df.Purchased)
plt.title("Before")
plt.show()

sns.scatterplot(x=x1_train.Age,y=x1_train.EstimatedSalary,hue=df.Purchased)
plt.title("After")
plt.show()
```
